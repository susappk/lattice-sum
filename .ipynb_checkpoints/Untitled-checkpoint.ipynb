# a program used to caculate the extinction crosssection of vortex Au particle

## 1. import n k material

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
wavelength_nm,exp_n,exp_k,fdtd_n,fdtd_k=np.loadtxt("/Users/zhongguoxia/Desktop/Au (Gold) - Palik.txt",unpack='true')
lam=wavelength_nm
n=fdtd_n+1j*fdtd_k
k_0=2*np.pi/(lam*1e-9)
k_s=k_0

d=400e-9  #axis direction lattice period
c=300e-9  #length of the edge of triangle
e=100e-9  #axis direction initial position distance  
lat=401 #sum samples in the axis direction
```

## 2. 1st order sphercal bessel function

```python
def j(x):
    return np.sin(x)/x**2-np.cos(x)/x
def jj(x):
    return np.sin(x)+np.cos(x)/x-np.sin(x)/x**2        
def h(x):
    return np.exp(1j*x)*(-1/x-1j/x**2)        
def hh(x):
    return -1j*np.exp(1j*x)+(x+1j)*np.exp(1j*x)/x**2
def alphap(a):
    rho =k_0*a*1e-9#半径????       
    a1=(n**2*j(n*rho)*jj(rho)-j(rho)*jj(n*rho))/(n**2*j(n*rho)*hh(rho)-h(rho)*jj(n*rho))
    return 1j*6*np.pi*a1/k_0**3
```

```python
plt.figure(1)
plt.plot(lam,k_0*np.imag(alphap(50)),label="$a$")
plt.xlabel("$lambda$")
plt.ylabel("$\sigma_{ext}$")
plt.title("test one particle mie-extinction")
plt.legend()
plt.show()
```

## 3.propagater function

 \begin{eqnarray}
\hat{G}(\vec{r})=\left((1+\frac{i}{k_s\vert\vec{r}\vert} - \frac{1}{k_{s}^{2}\vert \vec{r} \vert^{2}})\hat{u}+(-1-\frac{i3}{k_s\vert\vec{r}\vert} +\frac{3}{k_{s}^{2}\vert \vec{r} \vert^{2}})\vec{n}\vec{n}\right)\frac{e^{ik_{s}\vert \vec{r}\vert}}{4\pi\vert\vec{r}\vert}
\end{eqnarray}

direction of $\vec{r}$ is the direction of periodic ,dirction of incident light is $\vec{E}$

```python
def G(r,k_s):
    r=r.reshape(1,3)  
    R=np.linalg.norm(r)
        
    u=np.eye(3,dtype=np.complex)
    nn=r.T.dot(r)/R**2
       
    out=np.zeros((3,3),dtype=np.complex)    
    for i in np.arange(np.size(k_s)):
        A=(1+1j/k_s/R-1/k_s**2/R**2)
        B=(-1-1j*3/k_s/R+3/k_s**2/R**2)
        out=np.exp(1j*k_s*R)/R/np.pi/4*(A*u+B*nn)
    return out
```

## 4.sum matrix

the incident light:
\begin{eqnarray}
\vec{E}=E_{0}[1,i,0];
r_{a}-r_{a'}=nd\vec{z}  ;r_{a}-r_{b}=1\vec{x}+\sqrt{3}\vec{y}+(nd+150)\vec{z}  ;r_{a}-r_{b}=-1\vec{x}+\sqrt{3}\vec{y}+(nd+1)\vec{z}
\end{eqnarray}
add the omitted const
\begin{eqnarray}
 \left(\begin{array}{c}
 \vec{P}(r_{a})\\ \vec{P}(r_{b})\\\vec{P}(r_{c})
 \end{array}\right)=\left(\frac{\hat{I}}{\alpha_{p}}-\frac{k_{0}^{2}}{\epsilon_{0}}
 \left(\begin{array}{ccc}
 	 \sum_{a\not=a'}\hat{G}(r_{a},r_{a'})&\sum_{a}\hat{G}(r_{a},r_{b})&\sum_{a}\hat{G}(r_{a},r_{c})\\ 
        \sum_{b}\hat{G}(r_{b},r_{a})&\sum_{b\not=b'}\hat{G}(r_{b},r_{b'}) &\sum_{b}\hat{G}(r_{b},r_{c})\\ 
    	 \sum_{c}\hat{G}(r_{c},r_{a})&\sum_{c}\hat{G}(r_{c},r_{b})& \sum_{c\not=c'}\hat{G}(r_{c},r_{c'})\\ 
\end{array}\right)\right)^{-1}
 \left(\begin{array}{c}
  E\\ E\\ E
\end{array}\right)
\end{eqnarray}

```python
#I=np.eye(9,dtype=np.complex)
data=np.zeros((1000,lat,9,9),dtype=np.complex)
matrix=np.zeros((1000,9,9),dtype=np.complex)
result=np.zeros((1000,9),dtype=np.complex)
pa=np.zeros((1000),dtype=np.complex)
pb=np.zeros((1000),dtype=np.complex)
pc=np.zeros((1000),dtype=np.complex)

#E=np.array([1,1j,0,1,1j,0,1,1j,0])
for i in np.arange(1000):
    for m in np.arange(lat):
        f=m-(lat-1)/2
        ab=np.array([1*c,np.sqrt(3),f*d+e])
        ac=np.array([-1*c,np.sqrt(3),f*d+2*e])
        bc=np.array([-2*c,0,f*d+e])
        aa=np.array([0,0,f*d])
        
        if f==0:        
            data[i,m,0:3,3:6]=G(ab,k_s[i])
            data[i,m,0:3,6:9]=G(ac,k_s[i])
            
            data[i,m,3:6,0:3]=G(-ab,k_s[i])        
            data[i,m,3:6,6:9]=G(bc,k_s[i])
            
            data[i,m,6:9,0:3]=G(-ac,k_s[i])
            data[i,m,6:9,3:6]=G(-bc,k_s[i])
        else:
        
            data[i,m,0:3,0:3]=G(aa,k_s[i])
            data[i,m,3:6,3:6]=G(aa,k_s[i])
            data[i,m,6:9,6:9]=G(aa,k_s[i])
                
            data[i,m,0:3,3:6]=G(ab,k_s[i])
            data[i,m,0:3,6:9]=G(ac,k_s[i])
                
            data[i,m,3:6,0:3]=G(-ab,k_s[i])        
            data[i,m,3:6,6:9]=G(bc,k_s[i])
                
            data[i,m,6:9,0:3]=G(-ac,k_s[i])
            data[i,m,6:9,3:6]=G(-bc,k_s[i])
    
        matrix[i,:,:]=matrix[i,:,:]+data[i,m,:,:]
    result[i,:]=np.linalg.inv(I/alphap(50)[i]-k_0[i]**2*matrix[i,:,:]).dot(E)
#   pa[i]=k_0[i]*np.imag(np.array([1,1j,0]).dot(result[i,0:3]))
  #  pb[i]=k_0[i]*np.imag(np.array([1,1j,0]).dot(result[i,3:6]))
    #  pc[i]=k_0[i]*np.imag(np.array([1,1j,0]).dot(result[i,6:9]))
```
